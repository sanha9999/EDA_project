```python
# Shots_acc column 
df_tot["Shots_acc"] = round(df_tot["SOT"] / df_tot["Shots"], 2)
df_tot
```

```python
df_mct["Shots_acc"] = round(df_mct["SOT"] / df_mct["Shots"], 2)
df_mct
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_tot.sort_values("Shots_acc", ascending=False)[:5], x="Player", y="Shots_acc")
fig.set(title=" ", ylabel=" ")
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_mct.sort_values("Shots_acc", ascending=False)[:5], x="Player", y="Shots_acc")
fig.set(title=" ", ylabel=" ")
```
