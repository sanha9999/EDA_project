## Tottenham과 Man City로 data frame나누기

```python
df_tot = df_result[df_result["Team"]=="TOT"]
df_mct = df_result[df_result["Team"]=="MCT"]
```

```python
df_tot
```

## 슛 성공률

```python
# Shots_acc column을 생성
df_tot["Shots_acc"] = round(df_tot["SOT"] / df_tot["Shots"], 2)
df_tot
```

```python
df_mct["Shots_acc"] = round(df_mct["SOT"] / df_mct["Shots"], 2)
df_mct
```

```python
df_result["Shots_acc"] = df_result["SOT"] / df_result["Shots"]
```

```python
df_result
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_tot.sort_values("Shots_acc", ascending=False)[:5], x="Player", y="Shots_acc")
fig.set(title="토튼햄의 슛성공률", ylabel="슛 성공률")
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_mct.sort_values("Shots_acc", ascending=False)[:5], x="Player", y="Shots_acc")
fig.set(title="맨시티의 슛성공률", ylabel="슛 성공률")
```

## 패스 성공률

```python
df_tot["Pass_acc"] = df_tot["PA"] / df_tot["Pass"]
df_tot
```

```python
df_mct["Pass_acc"] = df_mct["PA"] / df_mct["Pass"]
df_mct
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_tot.sort_values("Pass_acc", ascending=False)[:5], x="Player", y="Pass_acc")
fig.set(title="토튼햄의 패스 성공률", ylabel="패스 성공률")
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_mct.sort_values("Pass_acc", ascending=False)[:5], x="Player", y="Pass_acc")
fig.set(title="맨시티의 패스 성공률", ylabel="패스 성공률")
```

## 태클 성공률

```python
df_tot["TK_acc"] = df_tot["TK_S"] / df_tot["TK"]
df_tot
```

```python
df_mct["TK_acc"] = df_mct["TK_S"] / df_mct["TK"]
df_mct
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_tot.sort_values("TK_acc", ascending=False)[:5], x="Player", y="TK_acc")
fig.set(title="토튼햄의 태클 성공률", ylabel="태클 성공률")
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_mct.sort_values("TK_acc", ascending=False)[:5], x="Player", y="TK_acc")
fig.set(title="맨시티의 태클 성공률", ylabel="태클 성공률")
```
