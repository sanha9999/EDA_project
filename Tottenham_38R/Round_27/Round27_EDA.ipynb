```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_result = pd.read_csv("Tottenham_Everton_Round27.csv",index_col = 0)
df_result.head()
```

```python
df_et = df_result[20:]
df_tot = df_result[:20]

df_tot = df_tot[df_tot['Rating']!=0.0].reset_index(drop=True)
df_et = df_et[df_et['Rating']!=0.0].reset_index(drop=True)
```

```python
plt.rcParams['font.size'] = 30
plt.figure(figsize=(20, 15))
plt.rcParams['font.family'] = 'NanumGothic'

word_df = df_tot.sort_values("Rating", ascending=False)[:5].reset_index(drop=True)

fig = sns.barplot(data=df_tot.sort_values("Rating", ascending=False)[:5], x="Player", y="Rating")
cnt = 0
for i in word_df["Rating"]:
    fig.text(x=cnt, y=i, s=i, color='black', ha="center",)
    cnt += 1
fig.set(title="토트넘 선수의 평점", ylabel="평점", xlabel="선수")
```

```python
plt.rcParams['font.size'] = 30
plt.figure(figsize=(20, 15))
plt.rcParams['font.family'] = 'NanumGothic'

word_df = df_tot.sort_values("Rating", ascending=True)[:5].reset_index(drop=True)

fig = sns.barplot(data=df_tot.sort_values("Rating", ascending=True)[:5], x="Player", y="Rating")
cnt = 0
for i in word_df["Rating"]:
    fig.text(x=cnt, y=i, s=i, color='black', ha="center",)
    cnt += 1
fig.set(title="오늘경기 worst", ylabel="평점", xlabel="선수")
```

```python
plt.rcParams['font.size'] = 30
plt.figure(figsize=(20, 15))
plt.rcParams['font.family'] = 'NanumGothic'

word_df = df_et.sort_values("Rating", ascending=False)[:5].reset_index(drop=True)

fig = sns.barplot(data=df_et.sort_values("Rating", ascending=False)[:5], x="Player", y="Rating")
cnt = 0
for i in word_df["Rating"]:
    fig.text(x=cnt, y=i, s=i, color='black', ha="center",)
    cnt += 1
fig.set(title="에버튼 선수의 평점", ylabel="평점", xlabel="선수")
```

```python
plt.rcParams['font.size'] = 30
plt.figure(figsize=(20, 15))
plt.rcParams['font.family'] = 'NanumGothic'

word_df = df_et.sort_values("Rating", ascending=True)[:5].reset_index(drop=True)

fig = sns.barplot(data=df_et.sort_values("Rating", ascending=True)[:5], x="Player", y="Rating")
cnt = 0
for i in word_df["Rating"]:
    fig.text(x=cnt, y=i, s=i, color='black', ha="center",)
    cnt += 1
fig.set(title="에버튼 worst", ylabel="평점", xlabel="선수")
```

```python
df_tot["team"] = "Tottenham"
df_et["team"] = "Everton"
df = pd.concat([df_tot, df_et], ignore_index=True)
df
```

```python
plt.rcParams['font.size'] = 30
plt.figure(figsize=(20, 15))
plt.rcParams['font.family'] = 'NanumGothic'

fig = sns.barplot(data=df, x="Position", y="Rating", hue='team')
fig.set(title="포지션 별 평점 비교", ylabel="평점", xlabel="포지션")
```
