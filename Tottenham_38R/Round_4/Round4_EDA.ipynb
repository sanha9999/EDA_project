```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_result = pd.read_csv("Crystal-Palace_Tottenham_Round4.csv",index_col = 0)
df_result.head()
```

```python
df_result.info()
```

```python
# 팀 구분
df_cp = df_result[:20]
df_tot = df_result[20:]
```

```python
df_cp = df_cp[df_cp['Rating']!=0.0].reset_index(drop=True)
df_cp
```

```python
df_tot = df_tot[df_tot['Rating']!=0.0].reset_index(drop=True)
df_tot
```

```python
plt.rcParams['font.size'] = 30
plt.figure(figsize=(20, 15))
plt.rcParams['font.family'] = 'NanumGothic'

word_df = df_tot.sort_values("Rating", ascending=False)[:5].reset_index(drop=True)

fig = sns.barplot(data=df_tot.sort_values("Rating", ascending=False)[:5], x="Player", y="Rating")
cnt = 0
for i in word_df["Rating"]:
    fig.text(x=cnt, y=i, s=i, color='black', ha="center",)
    cnt += 1
fig.set(title="토튼햄선수의 평점", ylabel="평점",)
```

```python
plt.figure(figsize=(20, 15))
plt.rcParams['font.size'] = 25

word_df = df_cp.sort_values("Rating", ascending=False)[:5].reset_index(drop=True)

fig = sns.barplot(data=df_cp.sort_values("Rating", ascending=False)[:5], x="Player", y="Rating")
cnt = 0
for i in word_df["Rating"]:
    fig.text(x=cnt, y=i, s=i, color='black', ha="center",)
    cnt += 1
fig.set(title="크리스탈 팰리스 선수의 평점", ylabel="평점",)
```

```python
df_tot["team"] = "Tottenham"
df_tot
```

```python
df_cp["team"] = "Crystal Palace"
df_cp
```

```python
df = pd.concat([df_tot, df_cp], ignore_index=True)
df
```

```python
plt.rcParams['figure.figsize'] = [20, 20]
plt.rcParams['font.size'] = 25
plt.rc('font', family='NanumGothic')

fig = sns.relplot(x="Pass", y="PA", hue="team", size="Rating", sizes=(100, 1000), alpha=.5, palette="muted",
            height=10, data=df)
fig.set(title="양 팀 패스 성공률", ylabel="성공한 패스", xlabel="시도한 패스")
```

```python
fig = sns.relplot(x="Shots", y="SOT", hue="team", size="Rating", sizes=(100, 1000), alpha=.5, palette="muted",
            height=10, data=df)
fig.set(title="양 팀 슛 성공률", ylabel="성공한 슛", xlabel="시도한 슛")
```

```python
fig = sns.relplot(x="DB", y="DB_W", hue="team", size="Rating", sizes=(100, 1000), alpha=.5, palette="muted",
            height=10, data=df)
fig.set(title="양 팀 드리블 성공률", ylabel="드리블 성공", xlabel="드리블 시도")
```
