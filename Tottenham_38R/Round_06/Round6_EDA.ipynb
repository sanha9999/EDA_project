```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_result = pd.read_csv("Tottenham_Arsenal_Round6.csv",index_col = 0)
df_result.head()
```

```python
df_as = df_result[:20]
df_tot = df_result[20:]

df_tot = df_tot[df_tot['Rating']!=0.0].reset_index(drop=True)
df_as = df_as[df_as['Rating']!=0.0].reset_index(drop=True)
```

```python
plt.rcParams['font.size'] = 30
plt.figure(figsize=(20, 15))
plt.rcParams['font.family'] = 'NanumGothic'

word_df = df_tot.sort_values("Rating", ascending=False)[:5].reset_index(drop=True)

fig = sns.barplot(data=df_tot.sort_values("Rating", ascending=False)[:5], x="Player", y="Rating")
cnt = 0
for i in word_df["Rating"]:
    fig.text(x=cnt, y=i, s=i, color='black', ha="center",)
    cnt += 1
fig.set(title="토트넘 선수의 평점", ylabel="평점", xlabel="선수")
```

```python
plt.rcParams['font.size'] = 30
plt.figure(figsize=(20, 15))
plt.rcParams['font.family'] = 'NanumGothic'

word_df = df_as.sort_values("Rating", ascending=False)[:5].reset_index(drop=True)

fig = sns.barplot(data=df_as.sort_values("Rating", ascending=False)[:5], x="Player", y="Rating")
cnt = 0
for i in word_df["Rating"]:
    fig.text(x=cnt, y=i, s=i, color='black', ha="center",)
    cnt += 1
fig.set(title="아스널 선수의 평점", ylabel="평점", xlabel="선수")
```

```python
df_tot["team"] = "Tottenham"
df_as["team"] = "Arsenal"
df = pd.concat([df_tot, df_as], ignore_index=True)
df
```

```python
plt.rcParams['font.size'] = 30
plt.figure(figsize=(20, 15))
plt.rcParams['font.family'] = 'NanumGothic'

fig = sns.barplot(data=df, x="Position", y="Rating", hue='team')
```

```python
plt.rcParams['font.size'] = 30
plt.figure(figsize=(20, 15))
plt.rcParams['font.family'] = 'NanumGothic'

fig = sns.barplot(data=df, x="Position", y="PA", hue='team')
fig.set(title="전체 포지션 패스 성공 횟수 비교", ylabel="패스 성공 횟수", xlabel="포지션")
```

```python
plt.rcParams['font.size'] = 30
plt.figure(figsize=(20, 15))
plt.rcParams['font.family'] = 'NanumGothic'

df_d = df[(df['Position']=='FWR') | (df['Position']=='FWL') | (df['Position']=='FW') | (df['Position']=='AML') | (df['Position']=='AMC') | (df['Position']=='AML')]

fig = sns.barplot(data=df_d, x="Position", y="Shots", hue='team')
fig.set(title="공격진 슈팅 비교", ylabel="슈팅 횟수", xlabel="포지션")
```
