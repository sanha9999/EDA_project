```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_result = pd.read_csv("data/round3.csv")
df_result
```

##   

```python
sns.barplot(data=df_result, x='Team', y='Rating')
```

##  

```python
df_tot = df_result[df_result["Team"]=="TOT"]
df_wat = df_result[df_result["Team"]=="WAT"]
```

```python
# Shots_acc column 
df_tot["Shots_acc"] = round(df_tot["SOT"] / df_tot["Shots"], 2)
df_tot
```

```python
df_wat["Shots_acc"] = round(df_wat["SOT"] / df_wat["Shots"], 2)
df_wat
```

```python
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['font.size'] = 20
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_tot.sort_values("Shots_acc", ascending=False)[:5], x="Player", y="Shots_acc")
fig.set(title=" ", ylabel=" ", xlabel=" ")
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_wat.sort_values("Shots_acc", ascending=False)[:5], x="Player", y="Shots_acc")
fig.set(title=" ", ylabel=" ", xlabel=" ")
```

```python
df_tot["Pass_acc"] = df_tot["PA"] / df_tot["Pass"]
df_tot
```

```python
df_wat["Pass_acc"] = df_wat["PA"] / df_wat["Pass"]
df_wat
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_tot.sort_values("Pass_acc", ascending=False)[:5], x="Player", y="Pass_acc")
fig.set(title="  ", xlabel=" ", ylabel=" ")
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_wat.sort_values("Pass_acc", ascending=False)[:5], x="Player", y="Pass_acc")
fig.set(title="  ", xlabel=" ", ylabel=" ")
```
