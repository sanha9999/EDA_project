```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_result = pd.read_csv("data/round2.csv")
df_result
```

```python
df_result.info()
```

## 양 팀 전체 평점

```python
sns.barplot(data=df_result, x='Team', y='Rating')
```

## 슛 성공률

```python
df_tot = df_result[df_result["Team"]=="TOT"]
df_wol = df_result[df_result["Team"]=="WOL"]
```

```python
# Shots_acc column을 생성
df_tot["Shots_acc"] = round(df_tot["SOT"] / df_tot["Shots"], 2)
df_tot
```

```python
df_wol["Shots_acc"] = round(df_wol["SOT"] / df_wol["Shots"], 2)
df_wol
```

```python
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['font.size'] = 20
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_tot.sort_values("Shots_acc", ascending=False)[:5], x="Player", y="Shots_acc")
fig.set(title="토튼햄의 슛성공률", ylabel="슛 성공률")
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_wol.sort_values("Shots_acc", ascending=False)[:5], x="Player", y="Shots_acc")
fig.set(title="울브스의 슛성공률", ylabel="슛 성공률", xlabel="선수")
```

## 패스 성공률

```python
df_tot["Pass_acc"] = df_tot["PA"] / df_tot["Pass"]
df_tot
```

```python
df_wol["Pass_acc"] = df_wol["PA"] / df_wol["Pass"]
df_wol
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_tot.sort_values("Pass_acc", ascending=False)[:5], x="Player", y="Pass_acc")
fig.set(title="토튼햄의 패스 성공률", xlabel="선수 이름", ylabel="패스 성공률")
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_wol.sort_values("Pass_acc", ascending=False)[:5], x="Player", y="Pass_acc")
fig.set(title="울브스의 패스 성공률", xlabel="선수 이름", ylabel="패스 성공률")
```
