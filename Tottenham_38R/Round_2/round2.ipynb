## 슛 성공률

```python
df_tot = df_result[df_result["Team"]=="TOT"]
df_wol = df_result[df_result["Team"]=="WOL"]
```

```python
# Shots_acc column을 생성
df_tot["Shots_acc"] = round(df_tot["SOT"] / df_tot["Shots"], 2)
df_tot
```

```python
df_wol["Shots_acc"] = round(df_wol["SOT"] / df_wol["Shots"], 2)
df_wol
```

```python
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['font.size'] = 20
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_tot.sort_values("Shots_acc", ascending=False)[:5], x="Player", y="Shots_acc")
fig.set(title="토튼햄의 슛성공률", ylabel="슛 성공률")
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_wol.sort_values("Shots_acc", ascending=False)[:5], x="Player", y="Shots_acc")
fig.set(title="울브스의 슛성공률", ylabel="슛 성공률", xlabel="선수")
```
