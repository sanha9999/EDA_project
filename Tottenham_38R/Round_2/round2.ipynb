```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_result = pd.read_csv("data/round2.csv")
df_result
```

```python
df_result.info()
```

##    

```python
sns.barplot(data=df_result, x='Team', y='Rating')
```

##  

```python
df_tot = df_result[df_result["Team"]=="TOT"]
df_wol = df_result[df_result["Team"]=="WOL"]
```

```python
# Shots_acc column 
df_tot["Shots_acc"] = round(df_tot["SOT"] / df_tot["Shots"], 2)
df_tot
```

```python
df_wol["Shots_acc"] = round(df_wol["SOT"] / df_wol["Shots"], 2)
df_wol
```

```python
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['font.size'] = 20
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_tot.sort_values("Shots_acc", ascending=False)[:5], x="Player", y="Shots_acc")
fig.set(title=" ", ylabel=" ")
```

```python
plt.figure(figsize=(20, 20))
fig = sns.barplot(data=df_wol.sort_values("Shots_acc", ascending=False)[:5], x="Player", y="Shots_acc")
fig.set(title=" ", ylabel=" ", xlabel="")
```
